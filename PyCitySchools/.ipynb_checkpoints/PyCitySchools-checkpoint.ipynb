```python
import pandas as pd
import numpy as np
```

```python
schools_path = "..//Resources//schools_complete.csv"
students_path = "..//Resources//students_complete.csv"
```

```python
#Read in data
schools_df = pd.read_csv(schools_path)
students_df = pd.read_csv(students_path)
```

```python
#Merge data
school_data_complete = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
#Ensure no bad values or blanks
school_data_complete.count()
```

```python
#Create series and values for District Summary
Total_Schools = len(schools_df['school_name'].unique())
Total_Students = students_df['student_name'].count()
Total_Budget = schools_df['budget'].sum()
Average_Math_Score = students_df['math_score'].mean()
Average_Reading_Score = students_df['reading_score'].mean()
Percent_Passing_Math = (len(students_df.loc[students_df['math_score'] >= 70]) / Total_Students) * 100
Percent_Passing_Reading = (len(students_df.loc[students_df['reading_score'] >= 70]) / Total_Students) * 100
Overall_Passing = (len(students_df.loc[(students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)]) / Total_Students) * 100

#Create District Summary table
District_Summary = pd.DataFrame({'Total Schools': [Total_Schools], 'Total Students': [Total_Students], 'Total Budget': [Total_Budget], 'Average Math Score': [Average_Math_Score], 'Average Reading Score': [Average_Reading_Score], '% Passing Math': [Percent_Passing_Math], '% Passing Reading': [Percent_Passing_Reading], '% Overall Passing': [Overall_Passing]})
District_Summary['Total Budget'] = District_Summary['Total Budget'].map("${0:,.2f}".format)
District_Summary['Average Math Score'] = District_Summary['Average Math Score'].map("{:.2f}".format)
District_Summary['Average Reading Score'] = District_Summary['Average Reading Score'].map("{:.2f}".format)
District_Summary['% Passing Math'] = District_Summary['% Passing Math'].map("{:.2f}%".format)
District_Summary['% Passing Reading'] = District_Summary['% Passing Reading'].map("{:.2f}%".format)
District_Summary['% Overall Passing'] = District_Summary['% Overall Passing'].map("{:.2f}%".format)

District_Summary
```

```python

#Create DataFrames of Passing Students
passing_math_df = school_data_complete[school_data_complete['math_score'] >= 70]
passing_reading_df = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_both_df = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]

#Creating Series and Values for School Summary Table
grouped_school = school_data_complete.groupby(['school_name'])
school_type = grouped_school['type'].unique().str[0]
school_students = grouped_school['Student ID'].count()
school_budget = grouped_school['budget'].unique().astype(int)
per_student_budget = school_budget / school_students
averageschool_math_score = grouped_school['math_score'].mean()
averageschool_reading_score = grouped_school['reading_score'].mean()
passing_math_schoolgrouped = passing_math_df.groupby(['school_name'])
passing_math_school = (passing_math_schoolgrouped['math_score'].count() / school_students) * 100
passing_reading_schoolgrouped = passing_reading_df.groupby(['school_name'])
passing_reading_school = (passing_reading_schoolgrouped['reading_score'].count() / school_students) * 100
passing_both_schoolgrouped = passing_both_df.groupby(['school_name'])
passing_both_school = (passing_both_schoolgrouped['student_name'].count() / school_students) * 100

#Create School Summary Table
school_summary_df = pd.DataFrame({'School Type': school_type, 'Total Students': school_students, 'Total School Budget': school_budget, 'Per Student Budget': per_student_budget, 'Average Math Score': averageschool_math_score, 'Average Reading Score': averageschool_reading_score, '% Passing Math': passing_math_school, '% Passing Reading': passing_reading_school, '% Overall Passing': passing_both_school})
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map("${0:,.2f}".format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map("${0:,.2f}".format)
school_summary_df['% Passing Math'] = school_summary_df['% Passing Math'].map("{0:,.2f}%".format)
school_summary_df['% Passing Reading'] = school_summary_df['% Passing Reading'].map("{0:,.2f}%".format)
school_summary_df['% Overall Passing'] = school_summary_df['% Overall Passing'].map("{0:,.2f}%".format)

school_summary_df
```

```python
top_schools_df = school_summary_df.sort_values('% Overall Passing', ascending=False)
top_schools_df.head(5)
```

```python
bottom_schools_df = school_summary_df.sort_values('% Overall Passing', ascending=True)
bottom_schools_df.head(5)
```

```python
#Create grade DF
grade9 = school_data_complete.loc[school_data_complete['grade'] == '9th']
grade10 = school_data_complete.loc[school_data_complete['grade'] == '10th']
grade11 = school_data_complete.loc[school_data_complete['grade'] == '11th']
grade12 = school_data_complete.loc[school_data_complete['grade'] == '12th']

school_grade9 = grade9.groupby(['school_name'])
school_grade10 = grade10.groupby(['school_name'])
school_grade11 = grade11.groupby(['school_name'])
school_grade12 = grade12.groupby(['school_name'])

#Grade Math Table
math_scores_by_grade = pd.DataFrame({'9th': school_grade9['math_score'].mean(), '10th': school_grade10['math_score'].mean(), '11th': school_grade11['math_score'].mean(), '12th': school_grade12['math_score'].mean()})
math_scores_by_grade
```

```python
#Grade Reading Table
reading_scores_by_grade = pd.DataFrame({'9th': school_grade9['reading_score'].mean(), '10th': school_grade10['reading_score'].mean(), '11th': school_grade11['reading_score'].mean(), '12th': school_grade12['reading_score'].mean()})
reading_scores_by_grade
```

```python
#Create per Student Budget Bins and Series
school_data_complete['student_budget'] = school_data_complete['budget'] / school_data_complete['size']
bins = [0, 585, 605, 625, 645, 665]
spending_groups = ["up to $585", "$586-605", "$606-625", "$626-645", "over $646"]
school_data_complete['Spending Ranges (Per Student)'] = pd.cut(school_data_complete['student_budget'], bins, labels=spending_groups)

school_budget_group = school_data_complete.groupby('Spending Ranges (Per Student)')

#Create per Student Values and Series
bybudget_students = school_budget_group['Student ID'].count()
passing_math_budget = school_data_complete[school_data_complete['math_score'] >= 70]
passing_math_budget_grouped = passing_math_budget.groupby(['Spending Ranges (Per Student)'])
passing_reading_budget = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_reading_budget_grouped = passing_reading_budget.groupby(['Spending Ranges (Per Student)'])
passing_both_budget = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
passing_both_budget_grouped = passing_both_budget.groupby(['Spending Ranges (Per Student)'])

passing_math_budgetcalc = (passing_math_budget_grouped['math_score'].count() / bybudget_students) * 100
passing_reading_budgetcalc = (passing_reading_budget_grouped['reading_score'].count() / bybudget_students) * 100
passing_both_budgetcalc = (passing_both_budget_grouped['Spending Ranges (Per Student)'].count() / bybudget_students) * 100

#Create Table from the Values
spending_range_df = pd.DataFrame({'Average Math Score': school_budget_group['math_score'].mean(), 'Average Reading Score': school_budget_group['reading_score'].mean(), '% Passing Math': passing_math_budgetcalc, '% Passing Reading': passing_reading_budgetcalc, '% Overall Passing': passing_both_budgetcalc})
spending_range_df['Average Math Score'] = spending_range_df['Average Math Score'].map("{0:,.2f}".format)
spending_range_df['Average Reading Score'] = spending_range_df['Average Reading Score'].map("{0:,.2f}".format)
spending_range_df['% Passing Math'] = spending_range_df['% Passing Math'].map("{0:,.2f}%".format)
spending_range_df['% Passing Reading'] = spending_range_df['% Passing Reading'].map("{0:,.2f}%".format)
spending_range_df['% Overall Passing'] = spending_range_df['% Overall Passing'].map("{0:,.2f}%".format)

spending_range_df
```

```python
#Create School size bins and Series
bins2 = [0, 1000, 2000, 5000]
size_groups = ["Small (<1000)", "Medium (1001-2000)", "Large (2001-5000)"]
school_data_complete['School Size Range'] = pd.cut(school_data_complete['size'], bins2, labels=size_groups)

school_size_group = school_data_complete.groupby('School Size Range')

#Create Series and Values
bysize_students = school_size_group['Student ID'].count()
passing_math_size = school_data_complete[school_data_complete['math_score'] >= 70]
passing_math_size_grouped = passing_math_size.groupby(['School Size Range'])
passing_reading_size = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_reading_size_grouped = passing_reading_size.groupby(['School Size Range'])
passing_both_size = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
passing_both_size_grouped = passing_both_size.groupby(['School Size Range'])

passing_math_sizecalc = (passing_math_size_grouped['math_score'].count() / bysize_students) * 100
passing_reading_sizecalc = (passing_reading_size_grouped['reading_score'].count() / bysize_students) * 100
passing_both_sizecalc = (passing_both_size_grouped['Spending Ranges (Per Student)'].count() / bysize_students) * 100

#Create School Size Summary Table
size_range_df = pd.DataFrame({'Average Math Score': school_size_group['math_score'].mean(), 'Average Reading Score': school_size_group['reading_score'].mean(), '% Passing Math': passing_math_sizecalc, '% Passing Reading': passing_reading_sizecalc, '% Overall Passing': passing_both_sizecalc})
size_range_df['Average Math Score'] = size_range_df['Average Math Score'].map("{0:,.2f}".format)
size_range_df['Average Reading Score'] = size_range_df['Average Reading Score'].map("{0:,.2f}".format)
size_range_df['% Passing Math'] = size_range_df['% Passing Math'].map("{0:,.2f}%".format)
size_range_df['% Passing Reading'] = size_range_df['% Passing Reading'].map("{0:,.2f}%".format)
size_range_df['% Overall Passing'] = size_range_df['% Overall Passing'].map("{0:,.2f}%".format)

size_range_df
```

```python
#Group by School Type
school_type_group = school_data_complete.groupby('type')

#Create Series and Values
bytype_students = school_type_group['Student ID'].count()
passing_math_type = school_data_complete[school_data_complete['math_score'] >= 70]
passing_math_type_grouped = passing_math_type.groupby(['type'])
passing_reading_type = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_reading_type_grouped = passing_reading_type.groupby(['type'])
passing_both_type = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
passing_both_type_grouped = passing_both_type.groupby(['type'])

passing_math_typecalc = (passing_math_type_grouped['math_score'].count() / bytype_students) * 100
passing_reading_typecalc = (passing_reading_type_grouped['reading_score'].count() / bytype_students) * 100
passing_both_typecalc = (passing_both_type_grouped['type'].count() / bytype_students) * 100

#Create summary table by school type
type_range_df = pd.DataFrame({'Average Math Score': school_type_group['math_score'].mean(), 'Average Reading Score': school_type_group['reading_score'].mean(), '% Passing Math': passing_math_typecalc, '% Passing Reading': passing_reading_typecalc, '% Overall Passing': passing_both_typecalc})
type_range_df['Average Math Score'] = type_range_df['Average Math Score'].map("{0:,.2f}".format)
type_range_df['Average Reading Score'] = type_range_df['Average Reading Score'].map("{0:,.2f}".format)
type_range_df['% Passing Math'] = type_range_df['% Passing Math'].map("{0:,.2f}%".format)
type_range_df['% Passing Reading'] = type_range_df['% Passing Reading'].map("{0:,.2f}%".format)
type_range_df['% Overall Passing'] = type_range_df['% Overall Passing'].map("{0:,.2f}%".format)

type_range_df
```
